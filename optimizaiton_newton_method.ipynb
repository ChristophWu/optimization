```python
import numpy as np
```

```python
def G(x):
    grad = np.zeros(4).reshape(4,1)
    grad[0] = 2 * (x[0] + 10 * x[1]) + 40 * (x[0] - x[3]) ** 3
    grad[1] = 20 * (x[0] + 10 * x[1]) + 4 * (x[1] - 2 * x[2]) ** 3
    grad[2] = 10 * (x[2] - x[3]) - 8 * (x[1] - 2 * x[2]) ** 3
    grad[3] = -10 * (x[2] - x[3]) - 40 * (x[0] - x[3]) ** 3
    return grad
```

```python
x = np.array([3,-1,0,1]).reshape(4,1)
g = G(x)
print(g)
```

```python
def H(x):
    hes = np.zeros([4,4])
    hes[0][0] = 2 + 120 * (x[0] - x[3]) ** 2
    hes[0][1] = 20
    hes[0][3] = -120 * (x[0] - x[3]) ** 2
    hes[1][0] = 20
    hes[1][1] = 200 + 12 * (x[1] - 2 * x[2]) ** 2
    hes[1][2] = -24 * (x[1] - 2 * x[2]) ** 2
    hes[2][1] = -24 * (x[1] - 2 * x[2]) ** 2
    hes[2][2] = 10 + 48 * (x[1] - 2 * x[2]) ** 2
    hes[2][3] = -10
    hes[3][0] = -120 * (x[0] -x[3]) ** 2
    hes[3][2] = -10
    hes[3][3] = 10 + 120 * (x[0] - x[3]) ** 2
    return hes
```

```python
h = H(x)
h_inv = np.linalg.inv(h)
print(h)
print(h_inv)
```

```python
x = np.array([3,-1,0,1]).reshape(4,1)
for i in range(3):
    print('iteration',i+1)
    h = H(x)
    print('h = ',h)
    g = G(x)
    print('g = ',g)
    h_inv = np.linalg.inv(h)
    print('h_inv = ',h_inv)
    x = x - np.matmul(h_inv,g)
    print('x_new = ',x)
```
